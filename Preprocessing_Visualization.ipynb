```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bezdekIris1.csv', na_values=['#NAME?'])
```

```python
df.describe()
```

```python
print(df.head(5))
```

```python
#printing missing values
print(df.isnull().sum())
```

```python
# Let's see how many examples we have of each species
df["Type"].value_counts()
```

```python
correla = df.corr(method='pearson')
print(correla)
```

```python
#print the contigency table between PetalWidth and Type to find chisquare value.
new_test_df = pd.crosstab(index=df['PetalWidth'],columns=df['Type'],margins=True)
print(new_test_df)
```

```python
#convert PetalWidth into categorical data and then find the crosstab with "Type"
#define function to find chi-squared value 
```

```python
#group the data by the class
grouped_data = df.groupby(['Type'])
grouped_data['Type'].describe().unstack()
```

```python
# find 5-point summary
data = df['PetalLength']
quartiles = np.percentile(data, [25, 50, 75])
data_min, data_max = data.min(), data.max()
print("5-number summary for PetalLength")
print('Min: %.3f' % data_min)
print('Q1: %.3f' % quartiles[0])
print('Median: %.3f' % quartiles[1])
print('Q3: %.3f' % quartiles[2])
print('Max: %.3f' % data_max)
```

```python
#Measuring Central Tendancy
grouped_data['PetalLength'].mean().reset_index()
```

```python
grouped_data['PetalLength'].aggregate(np.mean).reset_index()
```

```python
grouped_data['PetalLength'].median().reset_index()
grouped_data['PetalLength'].aggregate(np.median).reset_index()
```

```python
#Measures of variability - Std Dev
grouped_data['PetalLength'].std().reset_index()
```

```python
#IQR
grouped_data['PetalLength'].quantile([.25, .5, .75]).unstack()
```

```python
#Variance
grouped_data['PetalLength'].var().reset_index()
```

# Visualization

```python
%matplotlib inline

# The first way we can plot things is using the .plot extension from Pandas dataframes
# We'll use this to make a scatterplot of the Iris features.
df.plot(kind="scatter", x="SepalLength", y="SepalWidth")
df.plot(kind="scatter", x="PetalLength", y="PetalWidth")
```

```python
#plot a barplot of the frequency distribution of a categorical feature 
type_count = df['Type'].value_counts()
sns.barplot(type_count.index, type_count.values, alpha=0.9)
```

```python

# We can also use the seaborn library to make a similar plot
# A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure
sns.jointplot(x="SepalLength", y="SepalWidth", data=df, size=5)
```

```python
# One piece of information missing in the plots above is what species each plant is
# We'll use seaborn's FacetGrid to color the scatterplot by species
sns.FacetGrid(df, hue="Type", size=5) \
   .map(plt.scatter, "SepalLength", "SepalWidth") \
   .add_legend()
sns.FacetGrid(df, hue="Type", size=5) \
   .map(plt.scatter, "PetalLength", "PetalWidth") \
   .add_legend()
```

```python
df.boxplot("PetalLength", figsize=(12, 6))
```

```python
df.boxplot(by="Type", figsize=(12, 6))
```

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Type", y="PetalLength", data=df)
```

```python
# One way we can extend this plot is adding a layer of individual points on top of
# it through Seaborn's stripplot
# 
# We'll use jitter=True so that all the points don't fall in single vertical lines
# above the species
#
# Saving the resulting axes as ax each time causes the resulting plot to be shown
# on top of the previous axes
ax = sns.boxplot(x="Type", y="PetalLength", data=df)
ax = sns.stripplot(x="Type", y="PetalLength", data=df, jitter=True, edgecolor="gray")
```

```python

# A final seaborn plot useful for looking at univariate relations is the kdeplot,
# which creates and visualizes a kernel density estimate of the underlying feature
sns.FacetGrid(df, hue="Type", size=6) \
   .map(sns.kdeplot, "PetalLength") \
   .add_legend()
```

```python
# Another useful seaborn plot is the pairplot, which shows the bivariate relation
# between each pair of features
# 
# From the pairplot, we'll see that the Iris-setosa species is separataed from the other
# two across all feature combinations
sns.pairplot(df, hue="Type", size=3)
```

```python
# The diagonal elements in a pairplot show the histogram by default
# We can update these elements to show other things, such as a kde
sns.pairplot(df, hue="Type", size=3, diag_kind="kde")
```

```python
# Now that we've covered seaborn, let's go back to some of the ones we can make with Pandas
# We can quickly make a boxplot with Pandas on each feature split out by species
df.boxplot(by="Type", figsize=(12, 6))
```

```python
# Another multivariate visualization technique pandas has is parallel_coordinates
# Parallel coordinates plots each feature on a separate column & then draws lines
# connecting the features for each data sample
from pandas.tools.plotting import parallel_coordinates
parallel_coordinates(df, "Type")
```
